```python
import pandas as pd
import json
```

```python
with open("../../../preprocessing/preprocessed_data/tandler_upstream_gebiete.json", "r") as f:
    tandler_upstream_gebiete = json.load(f)
```

```python
import pandas as pd

chunksize = 1_000_000  # tune based on RAM

filtered_chunks = []

for chunk in pd.read_csv(
    "../../../preprocessing/data/num_agents_area_ag.txt",
    chunksize=chunksize,
):
    filtered = chunk[(chunk["t"] == 0) & (chunk["Sim"]==1)]
    if not filtered.empty:
        filtered_chunks.append(filtered)

df = pd.concat(filtered_chunks, ignore_index=True)
```

```python
agents_per_ww_area = df.groupby("area")["NumAgents"].sum()
```

```python
agents_upstream = {}

for sampling_station in tandler_upstream_gebiete:
    areas = tandler_upstream_gebiete[sampling_station]
    n_agents = 0
    for area in areas:
        n_agents += agents_per_ww_area[int(area)]
    agents_upstream[sampling_station] = n_agents
    print(f"Sampling station {sampling_station} has {n_agents} agents upstream.")
```
