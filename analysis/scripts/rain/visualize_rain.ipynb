```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys

sys.path.append(os.path.join(os.getcwd(), ".."))
from plotting_tools import * 
```

```python
df = pd.read_csv("rain.txt", decimal=",", sep=";", header=None, skiprows=1).reset_index()
df.columns = ["date", "rain"]
df['date'] = pd.date_range(start="2020-03-02 00:00", periods=len(df), freq="5min")
```

```python
df.describe()
```

```python
fig, ax = plt.subplots(figsize=(14, 3.5))
sns.lineplot(data=df, x="date", y="rain", ax=ax)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%d.%m."))  # Format without year
plt.tight_layout()
plt.ylim(0,90)
plt.ylabel("Precipitation [mm]")
plt.xlabel("Date")
plt.savefig("../../plots/rain.png")
plt.show()
```
